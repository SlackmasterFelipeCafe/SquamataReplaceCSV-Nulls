## Please set directory paths below

```python
#Set Data Paths - perhaps we'll get a user interface GUI to do this some day so the user doesn't have to change the code?
csvPath = r"C:\CurrentWork\DataReleases\IronMountain" #The 'r' signifies a string literal. Use this syntax for paths.
csvFileName = "USA-IronMtnChatham-AirborneMagnetics-2018-linedataOLD.csv"
```

```python
#Concatenate strings and check path for the fun of it.
csvFilePath = csvPath + "\\" + csvFileName
print ('The CSV File Path is ' + csvFilePath)
```

## Lets now import and index values from the CSV file

```python
#!!!!!!!!!!!!!!!!!!!!! PLEASE BE PATIENT !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# read file into pandas; note that can take over 1 minute depending on the size of the CSV file and computing power.
#for more on pandas visit https://pandas.pydata.org/
print ('!!!!!!!PLEASE BE PATIENT!!!!!!\nThis may take several minutes depending on the file size...')
csvDataFrame = pd.read_csv(csvFilePath, sep=',', delimiter=None, header='infer', names=None, index_col=None, usecols=None, squeeze=False, prefix=None, mangle_dupe_cols=True, dtype=None, engine=None, converters=None, true_values=None, false_values=None, skipinitialspace=False, skiprows=None, nrows=None, na_values=None, keep_default_na=True, na_filter=True, verbose=False, skip_blank_lines=True, parse_dates=False, infer_datetime_format=False, keep_date_col=False, date_parser=None, dayfirst=False, iterator=False, chunksize=None, compression='infer', thousands=None, decimal=b'.', lineterminator=None, quotechar='"', quoting=0, doublequote=True, escapechar=None, comment=None, encoding=None, dialect=None, tupleize_cols=None, error_bad_lines=True, warn_bad_lines=True, delim_whitespace=False, low_memory=True, memory_map=False, float_precision=None)
csvDataFrame
```

```python
#Check to see if there are any nulls.  If no nulls are present in the data set, the answer is False
#If the answer is 'True', run the script below to replace all null values in the data set with a user selected value.
csvDataFrame.isnull().values.any()
```

```python
# replacing na values in pandas data frame with a user selected value
#########################################################################################################################
newNullVal = '-999999999'
#########################################################################################################################
print ('!!!!!!!PLEASE BE PATIENT!!!!!!\nThis may take several minutes depending on the file size...')
newNullDataFrame = csvDataFrame.fillna(newNullVal) 
newNullDataFrame
```

### Now export the file as a csv with a differnt file name

```python
#Now export the new file with the null value changes in UTF-8 format
print ('!!!!!!!PLEASE BE PATIENT!!!!!!\nThis may take several minutes depending on the file size...')
newFileName = 'USA-IronMtnChatham-AirborneMagnetics-2018-linedataFINALCORRECTED.csv'
newFilePath = csvPath + "\\" + newFileName
newNullDataFrame.to_csv(newFilePath, index=False, encoding='utf-8')
print('New CSV file exported to:' + newFilePath)
```
